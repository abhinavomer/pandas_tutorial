```python
import pandas as pd
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','C','D'],'Class':[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({'ID':[7,8],'Name':['G','H'],'Class':[11,12]})
df2
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1,join='inner')
```

```python
pd.concat([df1,df2],axis=1,join='outer')
```

```python
pd.concat([df1,df2],axis=1,join_axes=[df1.index]) #Removed from current version
```

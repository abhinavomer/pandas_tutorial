```python
import pandas as pd
```

```python
"""Melt Function"""
```

```python
df=pd.read_csv('movie_rating.csv')
df
```

```python
pd.melt(df)
```

```python
pd.melt(df,id_vars=['Genre'])
```

```python
pd.melt(df,id_vars=['Year'])
```

```python
pd.melt(df,id_vars=['Year'],value_vars=['Genre'])
```

```python
pd.melt(df,id_vars=['Year'],value_vars=['Ratings %'])
```

```python
pd.melt(df,id_vars=['Year'],value_vars=['Genre'],var_name='Genre/Category')
```

```python
pd.melt(df,id_vars=['Year'],value_vars=['Genre'],var_name='Genre/Category',value_name='type')
```
